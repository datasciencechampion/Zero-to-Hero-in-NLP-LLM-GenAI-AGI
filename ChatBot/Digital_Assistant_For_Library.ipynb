## Defining System Prompt

```python
# System prompt with details about the books
system_prompt ="""
You are a Digital Assistant in a Library., an automated service to assist users about books available in this library.
Use the below library data while assisting user for their query.

Library Data:
Book Name,Author,Location(bookself number)
"The Alchemist",Paulo Coelho,A1
"Atomic Habits",James Clear,A2
"The Power of Now",Eckhart Tolle,A3
"Think and Grow Rich",Napoleon Hill,B1
"Drive: The Surprising Truth About What Motivates Us",Daniel H. Pink,B2
"Mindset: The New Psychology of Success",Carol S. Dweck,B3
"The 7 Habits of Highly Effective People",Stephen R. Covey,C1
"Man's Search for Meaning",Viktor E. Frankl,C2
"The Four Agreements",Don Miguel Ruiz,C3
"Awaken the Giant Within",Tony Robbins,D1

When a requested book is not available, collect the user phone number and tell them that we will callback when this book is available.

"""
```

## Define a Helper function for Digital Assistant
This helper function will accept user prompt and provide response based on the system prompt instruction.

```python
def get_assistance_response(messages):
    # Use the OpenAI API to generate a response
    response = client.chat.completions.create(
      model="gpt-3.5-turbo",
      messages=messages,
      temperature=0.2
    )
    return response.choices[0].message.content
```

```python
messages=[]
messages.append({"role": "system", "content": system_prompt})
```

```python
# Main loop for the user interface
while True:
    user_prompt = input("Ask me a question (or type 'exit' to quit):\n ")
    if user_prompt.lower() == 'exit':
        break
    messages.append({"role": "user", "content": user_prompt})
    # Get and print the response
    print("Assistant:", get_assistance_response(messages))
```
